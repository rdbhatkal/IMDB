## Load dataset

```python
#Importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
# Load in your csv.gz's of results for each year extracted
year2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
year2000.head()
```

### 2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
edadf['certification'].value_counts()
```

```python
## Lets look at Movies by Certification type

fig, ax = plt.subplots(figsize=(10, 8))
graph =sns.countplot( x ='certification',  data=edadf , palette='rainbow'); 


ax.set_title(' Count of Movies in each Certification Category', 
             fontfamily='serif',
             fontsize='xx-large',
             fontweight='semibold');

## Lets get the count

for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center");

## Increasing Axis Label Font Sizes
ax.set_xlabel('Certification Type',
              fontsize='x-large', fontweight='semibold')
ax.set_ylabel("Count",
              fontsize='x-large', fontweight='semibold');

```

* We have 124 movies in the PG-13 (Parents Strongly Cautioned) category 

* We have 154 movies in the R (Restricted) category 

* We have 28 movies in the PG (Parental Guidance Suggested) category 

* We have 10 movies in the G (General Audience) category 

* We have 4 movies in the NR (Not Rated) category 

### 3. What is the average revenue per certification category?

```python
## Lets look at Revenue of Movies by Certification type

## Creating a function with price ticks
from matplotlib.ticker import FuncFormatter
def million_M(x,pos):
  """function for use with matplotlib FuncFormatter - formats money in millions"""
  return f"${x*1e-6:,.0f}M"

# Create the formatter
price_fmt_mill =FuncFormatter(million_M)

# Lets plot
fig, ax = plt.subplots(figsize=(10, 8))
graph =sns.barplot( x ='certification', y = 'revenue',  data=edadf , palette='rainbow', ci = None); 


ax.set_title('Average Revenue for Movie by Certification Category', 
             fontfamily='serif',
             fontsize='xx-large',
             fontweight='semibold');

## Lets get the count

for p in graph.patches:
    height = round(p.get_height(),2)
    graph.annotate(million_M(height,0),(p.get_x()+p.get_width()/2, height + .05),ha="center")
    
    

## Increasing Axis Label Font Sizes
ax.set_xlabel('Certification Category',
              fontsize='x-large', fontweight='semibold')
ax.set_ylabel("Revenue",
              fontsize='x-large', fontweight='semibold');

## Now, use the ax.yaxis.set_major_formatter method 
ax.yaxis.set_major_formatter(price_fmt_mill)

```

* We have an average revenue of approximately $109 million in the PG-13 (Parents Strongly Cautioned) category

* We have an average revenue of approximately $54 million in the R (Restricted) category

* We have an average revenue of approximately $180 million in the PG (Parental Guidance Suggested) category

* We have an average revenue of approximately $178 million in the G (General Audience) category

* We have an average revenue of approximately $27 million in the NR (Not Rated) category

```python
## Lets look at budget for Movies by Certification type

# Lets plot
fig, ax = plt.subplots(figsize=(10, 8))
graph =sns.barplot( x ='certification', y = 'budget',  data=edadf , palette='rainbow', ci = None); 


ax.set_title('Average Budget for Movies by Certification Category', 
             fontfamily='serif',
             fontsize='xx-large',
             fontweight='semibold');

## Lets get the count

for p in graph.patches:
    height = round(p.get_height(),2)
    graph.annotate(million_M(height,0),(p.get_x()+p.get_width()/2, height + .05),ha="center")
    
    

## Increasing Axis Label Font Sizes
ax.set_xlabel('Certification Category',
              fontsize='x-large', fontweight='semibold')
ax.set_ylabel("Budget",
              fontsize='x-large', fontweight='semibold');

## Now, use the ax.yaxis.set_major_formatter method 
ax.yaxis.set_major_formatter(price_fmt_mill)

```
