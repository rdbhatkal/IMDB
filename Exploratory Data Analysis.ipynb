## Load dataset

```python
#Importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load in your csv.gz's of results for each year extracted
year2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
year2000.head()
```

```python
# Load in your csv.gz's of results for each year extracted
year2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
year2001.head()
```

```python
# concatenate the dataframes pages
newdf = pd.concat([year2001, year2001],
                      ignore_index=True)
```

```python
## Save current dataframe to file.
newdf.to_csv("Data/tmdb_results_combined.csv.gz",compression='gzip',index=False)
```
